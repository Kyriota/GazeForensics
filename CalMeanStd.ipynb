```python
from Core.TrainManager import TrainManager
from Config import Config
import numpy as np
```

```python
means, stds = [], []
for i in range(40):
    config = Config()
    trainManager = TrainManager(config)
    trainManager.current_DS_path = config.prep.tempDir + 'WDF_20.0%_epoch' + str(i) + '/'
    trainManager.init_data_loader()
    mean, std = trainManager.data_loader.dataset.get_normalization()
    print(mean, std)
    means.append(mean)
    stds.append(std)
```

```python
print(np.round([np.mean(np.array(means), axis=0), np.mean(np.array(stds), axis=0)], decimals=3))
```
